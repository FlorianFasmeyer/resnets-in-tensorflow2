```python
from Models import Resnets
import tensorflow as tf
```

# Alternating activation functions

```python
def custom_func(x): return tf.math.maximum(x, tf.math.sigmoid(x))

model = Resnets.cifar_resnet20(
        'preactivated', 
        shortcut_type='B', 
        activation=tf.keras.activations.swish, 
        activation2=custom_func,
        verbose=1)

tf.keras.utils.plot_model(model, show_shapes=True)
```

# Duplicating activation paths

```python
model = Resnets.cifar_resnet20(
        'dup_preactivated', 
        shortcut_type='B', 
        activation=tf.keras.activations.swish, 
        activation2=custom_func,
        pre_activation=tf.keras.activations.sigmoid,
        last_activation=tf.keras.activations.sigmoid,
        verbose=1)

tf.keras.utils.plot_model(model, show_shapes=True)
```
